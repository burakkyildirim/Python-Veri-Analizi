### TIPS VERİ SETİ

.total_bill: Yemeğe ödenen toplam tutar  
.tip: Bahşiş miktarı  
.sex: Ücreti ödeyen kişinin cinsiyeti (0= male, 1=female)  
.smoker: Sigara içip içmeme durumu (0=No, 1=Yes)  
.day: Gün(3=Thursday,4=Friday,5=Saturday,6=Sunday)  
.time: (0=Day, 1=Night)  
.size:Gruptaki kişi sayısı

. Hangi gün kazanç daha fazla?

```python
sns.boxplot(x=df["day"],y=df["total_bill"],data=df) # Pazar günü kazanç daha fazla
```

Hangi zaman diliminde kazanç daha fazla?

```python
sns.boxplot(x=df["time"],y=df["total_bill"],data=df) # Akşam kazanç daha fazla
```

Kişi sayısı ile kazanç arasındaki ilişki

```python
sns.boxplot(x=df["size"],y=df["total_bill"],data=df) # Kişi sayısı arttıkça kazançta artmaktadır.
```

Cinsiyet ile bahşiş miktarı arasında nasıl bir ilişki vardır?

```python
sns.boxplot(x=df["sex"],y=df["tip"],data=df) # Erkekler kadınlara göre daha fazla bahşiş vermiştir.
```

```python
df.groupby(df["sex"]).mean()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",style="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",palette="Set1", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time",palette="Set1", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time",row="sex",palette="Set1", data=df)
```
